## Function library

```python
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score
from scipy.stats import sem, t

def get_95_ci(data):
  '''
    Calculates the 95% confidence interval for a given set of data.

    Inputs
      data: data as a Pandas Series

    Outputs
      interval: Array of the lower and upper bounds of the confidence interval
  '''
  n = len(data)
  mean = data.mean()
  std_err = sem(data)
  interval = t.interval(0.95, n - 1, loc=mean, scale=std_err)
  return interval
```

## Performance and group fairness

```python
# BASELINE PERFORMANCE
accuracy_ci = get_95_ci(baseline_metrics['accuracy'])
roc_auc_ci = get_95_ci(baseline_metrics['roc_auc'])
fnr_ci = get_95_ci(baseline_metrics['fnr'])
fpr_ci = get_95_ci(baseline_metrics['fpr'])

# MODEL 0 PERFORMANCE
fair_0_accuracy_ci = get_95_ci(fair_0_metrics['accuracy'])
fair_0_roc_auc_ci = get_95_ci(fair_0_metrics['roc_auc'])
fair_0_fnr_ci = get_95_ci(fair_0_metrics['fnr'])
fair_0_fpr_ci = get_95_ci(fair_0_metrics['fpr'])

# MODEL 1 PERFORMANCE
fair_1_accuracy_ci = get_95_ci(fair_1_metrics['accuracy'])
fair_1_roc_auc_ci = get_95_ci(fair_1_metrics['roc_auc'])
fair_1_fnr_ci = get_95_ci(fair_1_metrics['fnr'])
fair_1_fpr_ci = get_95_ci(fair_1_metrics['fpr'])

# MODEL 2 PERFORMANCE
fair_2_accuracy_ci = get_95_ci(fair_2_metrics['accuracy'])
fair_2_roc_auc_ci = get_95_ci(fair_2_metrics['roc_auc'])
fair_2_fnr_ci = get_95_ci(fair_2_metrics['fnr'])
fair_2_fpr_ci = get_95_ci(fair_2_metrics['fpr'])

# MODEL 3 PERFORMANCE
fair_3_accuracy_ci = get_95_ci(fair_3_metrics['accuracy'])
fair_3_roc_auc_ci = get_95_ci(fair_3_metrics['roc_auc'])
fair_3_fnr_ci = get_95_ci(fair_3_metrics['fnr'])
fair_3_fpr_ci = get_95_ci(fair_3_metrics['fpr'])

perf_summary = pd.DataFrame({
    'Model': ['Baseline: Xdesc, Xcorr, Xind', 
              'Model 0: Udesc, Ucorr, Xind',
              'Model 1: Xdesc, Ucorr, Xind',
              'Model 2: Udesc, Xcorr, Xind',
              'Model 3: Udesc, Ucorr'],
    'Accuracy': [baseline_metrics['accuracy'].mean(),
                 fair_0_metrics['accuracy'].mean(),
                 fair_1_metrics['accuracy'].mean(),
                 fair_2_metrics['accuracy'].mean(),
                 fair_3_metrics['accuracy'].mean()
                 ],
    'Acc. 95% CI': [f'{round(accuracy_ci[0], 3)}-{round(accuracy_ci[1], 3)}',
                    f'{round(fair_0_accuracy_ci[0], 3)}-{round(fair_0_accuracy_ci[1], 3)}',
                    f'{round(fair_1_accuracy_ci[0], 3)}-{round(fair_1_accuracy_ci[1], 3)}',
                    f'{round(fair_2_accuracy_ci[0], 3)}-{round(fair_2_accuracy_ci[1], 3)}',
                    f'{round(fair_3_accuracy_ci[0], 3)}-{round(fair_3_accuracy_ci[1], 3)}'
                    ],
    'ROC AUC': [baseline_metrics['roc_auc'].mean(),
                fair_0_metrics['roc_auc'].mean(),
                fair_1_metrics['roc_auc'].mean(),
                fair_2_metrics['roc_auc'].mean(),
                fair_3_metrics['roc_auc'].mean()
                ],
    'ROC AUC 95% CI': [f'{round(roc_auc_ci[0], 3)}-{round(roc_auc_ci[1], 3)}',
                       f'{round(fair_0_roc_auc_ci[0], 3)}-{round(fair_0_roc_auc_ci[1], 3)}',
                       f'{round(fair_1_roc_auc_ci[0], 3)}-{round(fair_1_roc_auc_ci[1], 3)}',
                       f'{round(fair_2_roc_auc_ci[0], 3)}-{round(fair_2_roc_auc_ci[1], 3)}',
                       f'{round(fair_3_roc_auc_ci[0], 3)}-{round(fair_3_roc_auc_ci[1], 3)}'
                       ],
    'FNR': [baseline_metrics['fnr'].mean(),
            fair_0_metrics['fnr'].mean(),
            fair_1_metrics['fnr'].mean(),
            fair_2_metrics['fnr'].mean(),
            fair_3_metrics['fnr'].mean(),
            ],
    'FNR 95% CI': [f'{round(fnr_ci[0], 3)}-{round(fnr_ci[1], 3)}',
                  f'{round(fair_0_fnr_ci[0], 3)}-{round(fair_0_fnr_ci[1], 3)}',
                  f'{round(fair_1_fnr_ci[0], 3)}-{round(fair_1_fnr_ci[1], 3)}',
                  f'{round(fair_2_fnr_ci[0], 3)}-{round(fair_2_fnr_ci[1], 3)}',
                  f'{round(fair_3_fnr_ci[0], 3)}-{round(fair_3_fnr_ci[1], 3)}'
                  ],
    'FPR': [baseline_metrics['fpr'].mean(),
            fair_0_metrics['fpr'].mean(),
            fair_1_metrics['fpr'].mean(),
            fair_2_metrics['fpr'].mean(),
            fair_3_metrics['fpr'].mean()
            ],
    'FPR 95% CI': [f'{round(fpr_ci[0], 3)}-{round(fpr_ci[1], 3)}',
                  f'{round(fair_0_fpr_ci[0], 3)}-{round(fair_0_fpr_ci[1], 3)}',
                  f'{round(fair_1_fpr_ci[0], 3)}-{round(fair_1_fpr_ci[1], 3)}',
                  f'{round(fair_2_fpr_ci[0], 3)}-{round(fair_2_fpr_ci[1], 3)}',
                  f'{round(fair_3_fpr_ci[0], 3)}-{round(fair_3_fpr_ci[1], 3)}'
                  ]
})

print(perf_summary.to_markdown(index=False, numalign="right"))
```

```python
# Statistical significance of the difference in error rate disparity
baseline_metrics['fpr_diff'] = baseline_metrics['fpr_0'] - baseline_metrics['fpr_1']
baseline_metrics['fnr_diff'] = baseline_metrics['fnr_0'] - baseline_metrics['fnr_1']
fair_0_metrics['fpr_diff'] = fair_0_metrics['fpr_0'] - fair_0_metrics['fpr_1']
fair_0_metrics['fnr_diff'] = fair_0_metrics['fnr_0'] - fair_0_metrics['fnr_1']
fair_1_metrics['fpr_diff'] = fair_1_metrics['fpr_0'] - fair_1_metrics['fpr_1']
fair_1_metrics['fnr_diff'] = fair_1_metrics['fnr_0'] - fair_1_metrics['fnr_1']
fair_2_metrics['fpr_diff'] = fair_2_metrics['fpr_0'] - fair_2_metrics['fpr_1']
fair_2_metrics['fnr_diff'] = fair_2_metrics['fnr_0'] - fair_2_metrics['fnr_1']
fair_3_metrics['fpr_diff'] = fair_3_metrics['fpr_0'] - fair_3_metrics['fpr_1']
fair_3_metrics['fnr_diff'] = fair_3_metrics['fnr_0'] - fair_3_metrics['fnr_1']

fpr_fair_0_ttest = ttest_rel(baseline_metrics['fpr_diff'],
                           fair_0_metrics['fpr_diff'],
                           alternative='two-sided')
fnr_fair_0_ttest = ttest_rel(baseline_metrics['fnr_diff'],
                           fair_0_metrics['fnr_diff'],
                           alternative='two-sided')
fpr_fair_1_ttest = ttest_rel(baseline_metrics['fpr_diff'],
                           fair_1_metrics['fpr_diff'],
                           alternative='two-sided')
fnr_fair_1_ttest = ttest_rel(baseline_metrics['fnr_diff'],
                           fair_1_metrics['fnr_diff'],
                           alternative='two-sided')
fpr_fair_2_ttest = ttest_rel(baseline_metrics['fpr_diff'],
                           fair_2_metrics['fpr_diff'],
                           alternative='two-sided')
fnr_fair_2_ttest = ttest_rel(baseline_metrics['fnr_diff'],
                           fair_2_metrics['fnr_diff'],
                           alternative='two-sided')
fpr_fair_3_ttest = ttest_rel(baseline_metrics['fpr_diff'],
                           fair_3_metrics['fpr_diff'],
                           alternative='two-sided')
fnr_fair_3_ttest = ttest_rel(baseline_metrics['fnr_diff'],
                           fair_3_metrics['fnr_diff'],
                           alternative='two-sided')

print("FNR DISPARITY BETWEEN MALE AND FEMALE PATIENTS")
print(f"Average FNR Disparity in the baseline model:\
 {baseline_metrics['fnr_diff'].mean()*100:.2f}%")
print(f"Std Dev FNR Disparity in the baseline model:\
 {baseline_metrics['fnr_diff'].std()*100:.2f}%")
print('-'*20)
print(f"Average FNR Disparity in Model 0 (Udesc, Ucorr, Xind):\
 {fair_0_metrics['fnr_diff'].mean()*100:.2f}%")
print(f"Std Dev FNR Disparity in Model 0:\
 {fair_0_metrics['fnr_diff'].std()*100:.2f}%")
print(f"Paired T-Statistic (t): {fnr_fair_0_ttest.statistic:.4f}")
print(f"P-value: {fnr_fair_0_ttest.pvalue:.6f}")
print('-'*20)
print(f"Average FNR Disparity in Model 1 (Xdesc, Ucorr, Xind):\
 {fair_1_metrics['fnr_diff'].mean()*100:.2f}%")
print(f"Std Dev FNR Disparity in Model 1:\
 {fair_1_metrics['fnr_diff'].std()*100:.2f}%")
print(f"Paired T-Statistic (t): {fnr_fair_1_ttest.statistic:.4f}")
print(f"P-value: {fnr_fair_1_ttest.pvalue:.6f}")
print('-'*20)
print(f"Average FNR Disparity in Model 2 (Udesc, Xcorr, Xind):\
 {fair_2_metrics['fnr_diff'].mean()*100:.2f}%")
print(f"Std Dev FNR Disparity in Model 2:\
 {fair_2_metrics['fnr_diff'].std()*100:.2f}%")
print(f"Paired T-Statistic (t): {fnr_fair_2_ttest.statistic:.4f}")
print(f"P-value: {fnr_fair_2_ttest.pvalue:.6f}")
print('-'*20)
print(f"Average FNR Disparity in Model 3 (Udesc, Ucorr):\
 {fair_3_metrics['fnr_diff'].mean()*100:.2f}%")
print(f"Std Dev FNR Disparity in Model 3:\
 {fair_3_metrics['fnr_diff'].std()*100:.2f}%")
print(f"Paired T-Statistic (t): {fnr_fair_3_ttest.statistic:.4f}")
print(f"P-value: {fnr_fair_3_ttest.pvalue:.6f}")
print('-'*20)
```

```python
print("FPR DISPARITY BETWEEN MALE AND FEMALE PATIENTS")
print(f"Average FPR Disparity in the baseline model:\
 {baseline_metrics['fpr_diff'].mean()*100:.2f}%")
print(f"Std Dev FPR Disparity in the baseline model:\
 {baseline_metrics['fpr_diff'].std()*100:.2f}%")
print('-'*20)
print(f"Average FPR Disparity in Model 0 (Udesc, Ucorr, Xind):\
 {fair_0_metrics['fpr_diff'].mean()*100:.2f}%")
print(f"Std Dev FPR Disparity in Model 0:\
 {fair_0_metrics['fpr_diff'].std()*100:.2f}%")
print(f"Paired T-Statistic (t): {fpr_fair_0_ttest.statistic:.4f}")
print(f"P-value: {fpr_fair_0_ttest.pvalue:.6f}")
print('-'*20)
print(f"Average FPR Disparity in Model 1 (Xdesc, Ucorr, Xind):\
 {fair_1_metrics['fpr_diff'].mean()*100:.2f}%")
print(f"Std Dev FPR Disparity in Model 1:\
 {fair_1_metrics['fpr_diff'].std()*100:.2f}%")
print(f"Paired T-Statistic (t): {fpr_fair_1_ttest.statistic:.4f}")
print(f"P-value: {fpr_fair_1_ttest.pvalue:.6f}")
print('-'*20)
print(f"Average FPR Disparity in Model 2 (Udesc, Xcorr, Xind):\
 {fair_2_metrics['fpr_diff'].mean()*100:.2f}%")
print(f"Std Dev FPR Disparity in Model 2:\
 {fair_2_metrics['fpr_diff'].std()*100:.2f}%")
print(f"Paired T-Statistic (t): {fpr_fair_2_ttest.statistic:.4f}")
print(f"P-value: {fpr_fair_2_ttest.pvalue:.6f}")
print('-'*20)
print(f"Average FPR Disparity in Model 3 (Udesc, Ucorr):\
 {fair_3_metrics['fpr_diff'].mean()*100:.2f}%")
print(f"Std Dev FPR Disparity in Model 0:\
 {fair_3_metrics['fpr_diff'].std()*100:.2f}%")
print(f"Paired T-Statistic (t): {fpr_fair_3_ttest.statistic:.4f}")
print(f"P-value: {fpr_fair_3_ttest.pvalue:.6f}")
print('-'*20)
```

```python
# Visualising difference in FNR disparity between baseline and fair model
import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize=(4, 4))
sns.scatterplot(baseline_metrics, x='fnr_1', y='fnr_0', ax=ax, c='#1f77b4')
sns.scatterplot(fair_0_metrics, x='fnr_1', y='fnr_0', ax=ax, c='#ff7f0e')
plt.plot([0,.7],[0,.7],c='g')

fair_vx = (fair_0_metrics['fnr_1'] - baseline_metrics['fnr_1']).sum(axis=0) / baseline_metrics.shape[0]
fair_vy = (fair_0_metrics['fnr_0'] - baseline_metrics['fnr_0']).sum(axis=0) / baseline_metrics.shape[0]
fair_average_diff = mpatches.Arrow(0.05, 0.1, fair_vx, fair_vy, width=0.05, color="#ff7f0e" )
ax.add_patch(fair_average_diff)

fig.suptitle('False Negative Rate by subgroup')
plt.xlabel('Male FNR')
plt.ylabel('Female FNR')
plt.legend(labels=['Baseline model', 'Model 0',
                   'Equal error rates between subgroups',
                   'Average change in FNR from Baseline to Model 0'],
           loc='upper left', bbox_to_anchor=(1, 1))
ax.set_xlim([0,0.3])
ax.set_ylim([-.05,0.7])
ax.set_aspect('equal')
plt.show()
```
